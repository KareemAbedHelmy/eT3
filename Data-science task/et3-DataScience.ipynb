```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import pylab as pl
%matplotlib inline
```

```python
df=pd.read_csv('drinkMenu.csv')
```

```python
df.head()
```

```python
#Removed the beverage prep coloumn
df = df.drop(columns=['Beverage_prep'])
df.head()
```

```python
# this shows how many duplicates there are in the dataset
df.duplicated().sum()
```

```python
lenData = len(df)
#this function will drop any duplicate values present, this dataset had no duplicates so none were removed
df.drop_duplicates(keep = False, inplace = True)
lenDataAfter = len(df)
print(lenData,lenDataAfter)
```

```python
#any missing value will be filled with a 0
df.fillna(0)
```

```python
df.plot(x="Beverage", y="Calories", kind="bar", figsize=(50, 50))
```

```python
df.rename(columns = {'Sugars (g)':'Sugars'}, inplace = True)
```
